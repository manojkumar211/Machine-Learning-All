# Distributions:-

## 1. Uniform Distribution
- All outcomes are equally likely in a given range.
- Description: All outcomes are equally likely within a given range.
- Use Case: Random sampling or initialization of weights in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate uniform data
data = np.random.uniform(low=0.0, high=10.0, size=1000)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')
plt.title("Uniform Distribution")
plt.show()
```

## 2. Normal (Gaussian) Distribution
- The classic bell curve with mean mu and standard deviation sigma.
- Description: A bell-shaped curve where most values cluster around the mean.
- Use Case: Common in nature and statistics; used in regression, hypothesis testing, and error modeling.

```python
from scipy.stats import norm

# Generate normal data
mu, sigma = 0, 1  # mean and standard deviation
data = np.random.normal(mu, sigma, 1000)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.6, color='green')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Normal Distribution")
plt.show()
```

## 3. Binomial Distribution
- Counts the number of successes in n trials with success probability p.
- Description: Counts the number of successes in a fixed number of trials with a success probability p.
- Use Case: Modeling binary outcomes like coin tosses or yes/no experiments.

```python
from scipy.stats import binom

# Parameters
n, p = 10, 0.5
data = binom.rvs(n, p, size=1000)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.6, color='orange')
plt.title("Binomial Distribution")
plt.show()
```

## 4. Poisson Distribution
- Models the number of events in a fixed interval of time or space.
- Description: Models the number of events occurring in a fixed interval of time or space.
- Use Case: Event counts like customer arrivals or call center inquiries.

```python
from scipy.stats import poisson

# Parameters
mu = 3
data = poisson.rvs(mu, size=1000)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.6, color='purple')
plt.title("Poisson Distribution")
plt.show()
```

## 5. Exponential Distribution
- Models the time between events in a Poisson process.
- Description: Models the time between events in a Poisson process.
- Use Case: Time-to-failure analysis or inter-arrival times of events.

```python
from scipy.stats import expon

# Parameters
scale = 2.0  # 1/lambda
data = expon.rvs(scale=scale, size=1000)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.6, color='red')
plt.title("Exponential Distribution")
plt.show()
```

## 6. Bernoulli Distribution
- Models the outcome of a single trial (success or failure).
- Description: Models the outcome of a single trial with success probability p.
- Use Case: Basis for binomial distribution; used in binary classification problems.
